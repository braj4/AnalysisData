```python
import pandas as pd
import matplotlib.pyplot as plt

#------------this is my sample data-----------------
sales_data = pd.read_csv('Sales_April_2019.csv')

sales_data['Price Each'] = pd.to_numeric(sales_data['Price Each'], errors='coerce')
sales_data['Quantity Ordered'] = pd.to_numeric(sales_data['Quantity Ordered'], errors='coerce')

#---------------------here I calculate the total sales---------------
total_sales = (sales_data['Price Each'] * sales_data['Quantity Ordered']).sum()

#----------------------here I calculate the average price------------------------------------
average_price = sales_data['Price Each'].mean()

#----------------------Now calculate the total quantity ordered-------------------------------
total_quantity = sales_data['Quantity Ordered'].sum()

#-------------------------Calculate sales by product------------------------------
product_sales = sales_data.groupby('Product')['Price Each'].sum()

product_sales.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Sales by Product')
plt.show()

print(f'Total Sales: ${total_sales:.2f}')
print(f'Average Price: ${average_price:.2f}')
print(f'Total Quantity Ordered: {total_quantity}')
print('\nProduct Sales:')
print(product_sales)
```
